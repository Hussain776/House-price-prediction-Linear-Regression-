```python
'''Creating a function to generate a new feature called 'Luxury Home'
   with the help of 3 features ('Waterfront View','Condition of the House','Overall Grade')'''

def luxury_home(value):
    
    count=0
    
    if value[0] == 'Yes':
        count = count+1
    
    if value[1] in ['Excellent','Good']:
        count = count+1
    
    if value[2] > 7:
        count = count+1
        
    
    if count >= 2:
        return 'Yes'
    else:
        return 'No'   
```

```python
# Function to impute outliers of dependent (target variable) features

def outlier_imputer(a):
    
    q1=tar_var.quantile(0.25)
    q3=tar_var.quantile(0.75)
    
    iqr=q3-q1
    
    min_cap=q1-1.5*iqr
    max_cap=q3+1.5*iqr
    
    if a > max_cap:
        return(max_cap)
    if a < min_cap:
        return(min_cap)
    else:
        return(a)
```

```python
# imputing outliers in target variable

df['Sale Price']=df['Sale Price'].apply(outlier_imputer)
```

```python
# custom function for easy and efficient analysis of numerical univariate

def UVA_numeric(data, var_group):
  '''
  Univariate_Analysis_numeric
  takes a group of variables (INTEGER and FLOAT) and plot/print all the descriptives and properties along with KDE.

  Runs a loop: calculate all the descriptives of i(th) variable and plot/print it
  '''

  size = len(var_group)
  plt.figure(figsize = (5,5*size), dpi = 100)
    
  #looping for each variable
  for j,i in enumerate(var_group):
    
    # calculating descriptives of variable
    mini = data[i].min()
    maxi = data[i].max()
    ran = data[i].max()-data[i].min()
    mean = data[i].mean()
    median = data[i].median()
    st_dev = data[i].std()
    skew = data[i].skew()
    kurt = data[i].kurtosis()

    # calculating points of standard deviation
    points = mean-st_dev, mean+st_dev

    #Plotting the variable with every information
    plt.subplot(size,1,j+1)
    sns.kdeplot(data[i], shade=True)
    sns.lineplot(points, [0,0], color = 'white', label = "std_dev",)
    sns.scatterplot([mini,maxi], [0,0], color = 'orange', label = "min/max")
    sns.scatterplot([mean], [0], color = 'red', label = "mean")
    sns.scatterplot([median], [0], color = 'blue', label = "median")
    plt.xlabel('{}'.format(i), fontsize = 20)
    plt.ylabel('density')
    plt.title('std_dev = {}; kurtosis = {};\nskew = {}; range = {}\nmean = {}; median = {}'.format((round(points[0],2),round(points[1],2)),
                                                                                                   round(kurt,2),
                                                                                                   round(skew,2),
                                                                                                   (round(mini,2),round(maxi,2),round(ran,2)),
                                                                                                   round(mean,2),
                                                                                                   round(median,2)))
    plt.subplots_adjust(hspace=0.7)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_data = X

VIF = pd.Series([variance_inflation_factor(vif_data,i) for i in range(vif_data.shape[1])],index=vif_data.columns)
VIF
```

```python
vif_data
```

```python
#  Function to remove Multicollinearity

def MC_remover(data):
    vif = pd.Series([variance_inflation_factor(vif_data,i) for i in range(vif_data.shape[1])],index=data.columns)
    if vif.max() > 5:
        print(vif[vif == vif.max()].index[0], 'has been removed')
        data = vif_data.drop(columns = [vif[vif == vif.max()].index[0]])
        return data
    else:
        print('No Multicollinearity present anymore')
        return data
```

```python
for i in range(6):
    vif_data = MC_remover(vif_data)
    
vif_data.head()
```

# Forward Feature Selection & Backward Feature Elimination

### Forward Feature Selection

```python
# Function to calculate Adjusted R2

def adj_r2(m , data_y , predictions):
    
    from sklearn.metrics import r2_score
    R = r2_score(data_y, predictions)
    n = len(data_y)
    adj_r = 1 - ((1-R)*(n-1)/(n-m-1))
    return adj_r
```

```python
# Function to calculate R2 score after addition or removal of a variable

def new_adj_r2(X , Y , model):
    
    model.fit(X , Y)
    pred = model.predict(X)
    m = len(X.columns)
    r2 = adj_r2(m , Y , pred)
    return r2
```

```python
def forward_selection(model , X , Y , max_fea):
    
    f_col=[]
    for_hist=[]
    col=X
    
    for i in range(max_fea):
        f_col.append('blank')
        tmp=f_col[:]
        Rmax=0
        for var in col.columns:
            tmp[-1]=var
            R2 = new_adj_r2(X[tmp],Y,model)
            
            if R2 >= Rmax:
                f_col[-1]=var
                Rmax = R2
        for_hist.append(Rmax)
        col = col.drop(columns=f_col[-1],axis=1)
        print(i,'Selected ', f_col[-1] , 'with new r2 = ', Rmax)
    return Rmax , for_hist , X[f_col]
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression(normalize = True)

score, for_history, sel_data = forward_selection(lr, X_train, Y_train, 29)
```

```python
#  Plotting R2 values For Forward Feature Selection

plt.figure(figsize=(15,7),dpi=150)
plt.plot(range(1,len(for_history)+1,1),for_history,c='g')
plt.xlabel('No. of Columns')
plt.ylabel('R2 score')
plt.title('R2 score Summary')
```

### Backward feature Elimination

```python
def backward_elimination(model , X , Y , min_fea ):
    elim=X[:]
    back_hist=[]
    base_r2 =new_adj_r2(X,Y,model)
    print('Base R2 score is :  ', base_r2,'\n')
    
    
    for i in range(X.shape[1] - min_fea):
        
        difference = -100
        
        for var in elim.columns:
            tmp=elim.drop(columns=[var])
            r2 = new_adj_r2(tmp,Y,model)
            
            if (r2 - base_r2) > difference:
                difference = r2 - base_r2
                eliminate = var
                new_r2 = r2
        
        elim=elim.drop(columns=[eliminate],axis=1)
        back_hist.append(new_r2)
        print(i,eliminate , ' was eliminated New adjusted R2 = ' , new_r2)
        base_r2=new_r2
        
    return base_r2 , back_hist , elim
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression(normalize = True)

score , back_history, survided_data = backward_elimination(lr, X_train, Y_train, 1)
```

```python
#  Plotting R2 values For Backward Feature Selection

plt.figure(figsize=(15,7),dpi=150)
plt.plot(range(len(back_history),0,-1),back_history,c='r')
plt.xlabel('No. of Columns')
plt.ylabel('R2 score')
plt.title('R2 score Summary')
```
